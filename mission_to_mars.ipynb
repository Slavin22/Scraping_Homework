```python
# Dependencies
from bs4 import BeautifulSoup
import requests
from splinter import Browser
import pandas as pd

# URL to scrape
url = 'https://mars.nasa.gov/news/'

# Retrieve page with the requests module
html = requests.get(url)

# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(html.text, 'html.parser')
soup.body
```

```python
title = soup.find("div", class_="content_title")
title = title.find("a").text.strip()
title
```

```python
paragraph = soup.find("div", class_="rollover_description_inner").text.strip()
paragraph
```

```python
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path)
url2 = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
browser.visit(url2)
```

```python
browser.click_link_by_partial_text('FULL IMAGE')
```

```python
# Can't run until above click is performed
browser.click_link_by_partial_text('more info')
```

```python
url3 = browser.url
html3 = requests.get(url3)
soup3 = BeautifulSoup(html3.text, 'html.parser')
image = soup3.find("img", class_="main_image")["src"]
featured_image_url = "https://www.jpl.nasa.gov" + image
featured_image_url
```
